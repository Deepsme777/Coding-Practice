```python
n = int(input("Enter the number of students: "))

student_marks = {}

for i in range(n):
    name = input("Enter the name of student: ")
    score = float(input("Enter the score of student: "))
    if name in student_marks:
        student_marks[name].append(score)
    else:
        student_marks[name] = [score]

query_name = input("Enter the name of student to find the average: ")

marks = student_marks[query_name]
avg = sum(marks) / len(marks)


print(student_marks)
print(f"{avg:.2f}")
```

```python
from collections import OrderedDict

if __name__ == '__main__':
    n = int(input())
    prices = OrderedDict()

    for _ in range(n):
        item_name, price = input().split()
        price = int(price)

        prices[item_name] = prices.get(item_name, 0) + price

    for item_name, total_price in prices.items():
        print(item_name, total_price)
        print(f"The total price would be: {total_price}")
```

```python
from collections import OrderedDict

if __name__ == '__main__':
    n = int(input())
    prices = OrderedDict()

    for _ in range(n):

        line = input().rsplit(maxsplit=1)  # this will split the input into two parts:
                                           # the item name and the price, even if the item name contains spaces
        item_name = line[0]
        price = int(line[1])  


        if item_name in prices:
            prices[item_name] += price
        else:
            prices[item_name] = price

    for item_name, price in prices.items():
        total_price = price
        print(item_name, price)
        print( "The total price would be : " + str(price))
```

```python
from collections import deque

if __name__ == '__main__' :

    n = int(input())
    d = deque()

    for _ in range(n) :
        commands = input().split()

        cmd_type = commands[0]

        if cmd_type == 'append' :
            d.append(int(commands[1]))

        elif cmd_type == 'appendleft' :
            d.appendleft(int(commands[1]))

        elif cmd_type == 'pop' :
            d.pop()

        elif cmd_type == 'popleft' :
            d.popleft()

    print(*d)


# 6
# append 1
# append 2
# append 3
# appendleft 4
# pop
# popleft
```

```python
from collections import deque

t = int(input())

for _ in range(t) :

    n = int(input())
    cubes = deque(map(int, input().split()))

    top = float('inf')
    possible = True

    while cubes :
        left = cubes[0]
        right = cubes[-1]


        if left >= right and left <= top :
            top = cubes.popleft()

        elif right >= left and right <= top :
            top = cubes.pop()

        else :
            possible = False
            break

    print("Yes" if possible else "No")

    
```
