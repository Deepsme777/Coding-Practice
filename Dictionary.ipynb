```python
student = {
    "name": "Deepak",
    "age": 19,
    "city": "Erode"
}

print(student["name"])
```

```python

#  to add some values use update keys

student = {"name": "Deepak"}

student.update({"age": 19, "city": "Erode"})

print(student)
```

```python
# to find an avg values in an dictionary

n = int(input("Enter the number of students: "))

student_marks = {}

for i in range(n) :
    name = input("Enter the name of student: ")
    score = float(input("Enter the score of student: "))
    student_marks[name] = score

query_name = input("Enter the name of student to find the average: ")

marks = student_marks[query_name]

avg = sum(marks) / len(marks)


print(student_marks)
print(f"{avg:.2f}")
```
