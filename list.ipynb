```python
# heres is an list with inbuilt conditions 

N = int(input())

res = []

for i in range(N) :
    command = input().split()
    cmd_type = command[0]

    if cmd_type == "insert" :
        res.insert(int(command[1]) , int(command[2]))
            
    elif cmd_type == "print" :
        print(res)
            
    elif cmd_type == "remove" :
        res.remove(int(command[1]))
            
    elif cmd_type == "append" :
        res.append(int(command[1]))
            
    elif cmd_type == "sort" :
        res.sort()
            
    elif cmd_type == "pop" :
        res.pop()
            
    elif cmd_type == "reverse" :
        res.reverse()



        
```

```python
cars = ["Mclaren" , "Ferrari" , "BMW" , "Audi" , "Mercedes"]

for x in cars :
    print(x)

#  modify list items
cars[1] = "Lambogirhini"

print(cars)
for x in cars :
    print(x)

# add list items
cars.extend(["Porsche" , "Pagani"])
print(cars)


# remove items in list
cars.pop(3)
print(cars)

# for length in lists
length = len(cars)
print(length)

print(cars[1])
print(cars[1:3])

print("----------------------")

# for insert in ists
cars.insert(1, "BMW")
for x in cars :
    print(x)

cars.clear()
for x in cars :
    print(x)


print("Thse list is cleared now.")
print(cars)
```

```python
if __name__ == '__main__' :

    M = set(map(int , input().split()))
    
    print(M)

    N = set(map(int , input().split()))

    print(N)

    u = set(M).difference(set(N))
    v = set(N).difference(set(M))

    t = u.union(set(v))

    print(t)  # this wil reprsent the dictionary of the set difference between the two sets M and N.

    for i in sorted(t) :
        print(i)
```

```python
from itertools import combinations

if __name__ == '__main__' :
    S , k = input().split()

    S = S.upper()
    S = sorted(S)

    k = int(k)

    for i in range(1 , k+1) :
        for j in combinations(S , i) :
            print(''.join(j))
```
