```python
# about to add in an nested list ( number of lists [ to print the students name and grade of second scores if it has same score print their in an alphabetical order ] )


students = []

for i in range(int(input("Enter the number of students: "))) :
    name = input("Enter the name of student: ")
    score = float(input("Enter the score of student: "))
    students.append( name , score )

    score = sorted(list(set([x[1] for x in students])))

    second_highest = score[1]

    result = [x[0] for x in students if x[1] == second_highest]

    result.sort()
    for name in result :
        print(name)
```
