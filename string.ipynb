```python
# to cappitalize the lower case  and to lower the upper case Aquired from user input

def swap_case(s):
    string = ""

    for i in s:
        if i == i.upper():
            string = string + i.lower()
        
        elif i == i.lower():
            string = string + i.upper()

        else:
            string = string + i

        return string
    
swap_case("greetings from Co-Sync")

# under construction
```

```python
def split_and_join(line):
    line = line.split(" ")
    line = "-".join(line)
    return line
    

if __name__ == '__main__':
    line = input()
    result = split_and_join(line)
    print(result)

split_and_join("this is a string")

# this = this is a string   to this-is-a-string
```

```python
def mutate_string(string, position, character):
    position = int(position)
    p = position
    k = position + 1
    character = str(character)
    
    string = string[:p] + character + string[k:]
    return string

mutate_string("abracadabra", 5, "k")
```
