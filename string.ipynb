```python
# to cappitalize the lower case  and to lower the upper case Aquired from user input

def swap_case(s):
    string = ""

    for i in s:
        if i == i.upper():
            string = string + i.lower()
        
        elif i == i.lower():
            string = string + i.upper()

        else:
            string = string + i

        return string
    
swap_case("greetings from Co-Sync")

# under construction
```

```python
def mutate_string(string, position, character):
    position = int(position)
    p = position
    k = position + 1
    character = str(character)
    
    string = string[:p] + character + string[k:]
    return string

mutate_string("abracadabra", 5, "k")
```

```python
# to find no of times the sun_string occurs in the string

def count_substring(string, sub_string):
    count = 0 
    start = 0

    while True :
        pos = string.find(sub_string, start)
        if pos == -1 :
            break

        count = count + 1
        start = pos + 1
    return count

if __name__ == '__main__':
    string = " ABCDCDC"
    sub_string = "CDC"
    count = count_substring(string, sub_string)
    print(count)
    print( '"' + sub_string + '"')
```

```python
# You are given a string S and width w .
# Your task is to wrap the string into a paragraph of width w .

def wrap(string, max_width):
    if len(string) <= max_width:
        return string
    
    else:
        return string[0:max_width] + "\n" + wrap(string[max_width:], max_width)
    
wrap("ABCDEFGHIJKLIMNOQRSTUVWXYZ", 4)
```

```python
string,k = "AABCAAADA", 3

s = len(string)
    
for i in range(0, s, k):
    substring = string[i:i+k]
        

# this will remove the duplicate in character   
    
    seen = []
    for char in substring:
        if char not in seen:
            seen.append(char)
                
    print(''.join(seen))
```
