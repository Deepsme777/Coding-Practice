```python
def average(array):
    array = set(array)

    l = len(array)
    s = sum(array)

    avg = round((s / l),3)
    return avg

average([161 ,182, 161, 154, 176, 170, 167, 171, 170, 174])
```

```python
if __name__ == '__main__' :

    n , m = map(int, input().split())

    arr = list(map(int, input().split()))

    A = map(int, input().split())
    B = map(int, input().split())

    A = set(A)
    B = set(B)

    happiness = 0

    for x in arr: 

        if x in A :
            happiness += 1

        elif x in B :
            happiness -= 1

        else :
            pass

    print(happiness)


# this is an array of integers in happiness score.
# where A is a set of integers that will add 1 to the total happiness score
# B is a set of integers that will subtract 1 from the total happiness score.
# The function calculates the total happiness score based on the elements in the array and the sets A and B.

# the input is  

#  3 , 2
#  1 5 3
#  3 1
#  2 5
```
