```python
def average(array):
    array = set(array)

    l = len(array)
    s = sum(array)

    avg = round((s / l),3)
    return avg

average([161 ,182, 161, 154, 176, 170, 167, 171, 170, 174])
```

```python
if __name__ == '__main__':

    A = set(map(int, input().split()))
    B = set(map(int, input().split()))

    difference_A = A.difference(B)
    difference_B = B.difference(A)

    plus_difference = difference_A.union(difference_B)

    union = A.union(B)

    common = union.difference(plus_difference)
    print(common)
```

```python
from collections import deque

if __name__ == '__main__':

    n = int(input())
    d = deque()

    for i in range(n):
        commands = input()
        if 'append' in commands:
            d.append(int(commands.split()[1]))

        elif 'appendleft' in commands:
            d.appendleft(int(commands.split()[1]))

        elif 'pop' in commands:
            d.pop()

        elif 'popleft' in commands:
            d.popleft()
    print(*d)
```
