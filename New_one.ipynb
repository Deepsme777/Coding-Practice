```python
import re

def extract_numbers(text):
    # Use regular expression to find all numbers in the text
    numbers = re.findall(r'\d+', text)
    return numbers

# Example usage
text = "The price of the item is 100 dollars and the discount is 20% which makes it 80 dollars."
numbers = extract_numbers(text)
print(numbers)  # Output: ['100', '20', '80']
```

```python
def print_door_mat(N, M):
    # N is height (odd), M = 3*N is width
    
    # Top half (before center)
    for i in range(N // 2):
        pattern = '.|.' * (2 * i + 1)
        dashes = '-' * ((M - len(pattern)) // 2)
        print(dashes + pattern + dashes)
    
    # Center row with WELCOME
    welcome = 'WELCOME'
    dashes = '-' * ((M - len(welcome)) // 2)
    print(dashes + welcome + dashes)
    
    # Bottom half (after center - mirror of top)
    for i in range(N // 2 - 1, -1, -1):
        pattern = '.|.' * (2 * i + 1)
        dashes = '-' * ((M - len(pattern)) // 2)
        print(dashes + pattern + dashes)

# Example usage
N = 9  # Must be odd
M = 3 * N
print_door_mat(N, M)
```
