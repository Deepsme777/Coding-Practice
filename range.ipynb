```python
number = 17

width = len(bin(number)[2:])

for i in range(1,number+1):
    Decimal = str(i).rjust(width) # Decimal

    Octal = oct(i)[2:].rjust(width) # octal

    Hexa_decimal = hex(i)[2:].upper().rjust(width) # hexadecimal

    Binary = bin(i)[2:].rjust(width) # binary

    print(Decimal, Octal, Hexa_decimal, Binary)
```
